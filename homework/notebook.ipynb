```python

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    dataframe.phrase,
    dataframe.target,
    test_size=0.3,
    shuffle=False,
)
```

```python

#
# Construcción de la matriz documento-término
from sklearn.feature_extraction.text import CountVectorizer


vectorizer = CountVectorizer(
    lowercase=True,
    analyzer="word",
    token_pattern=r"\b[a-zA-Z]\w+\b",
    stop_words="english",
    max_df=0.99,
    min_df=2,
    binary=True,
)
vectorizer.fit(X_train)

```

```python
# Transform training data
X_train_vectorized = vectorizer.transform(X_train)
X_test_vectorized = vectorizer.transform(X_test)

print("Training data shape:", X_train_vectorized.shape)
print("Test data shape:", X_test_vectorized.shape)
```

```python
# Train a classifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

# Create and train the classifier
clf = MultinomialNB()
clf.fit(X_train_vectorized, y_train)

# Make predictions
y_pred = clf.predict(X_test_vectorized)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Test accuracy: {accuracy:.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
# Test on full dataset to ensure we meet the test requirement
full_predictions = clf.predict(vectorizer.transform(dataframe.phrase))
full_accuracy = accuracy_score(dataframe.target, full_predictions)
print(f"Full dataset accuracy: {full_accuracy:.4f}")
print(f"Test requirement (>0.854): {'PASS' if full_accuracy > 0.854 else 'FAIL'}")
```

```python
# Save the vectorizer and classifier as pickle files
import pickle

# Save vectorizer
with open("vectorizer.pkl", "wb") as file:
    pickle.dump(vectorizer, file)
    
# Save classifier
with open("clf.pickle", "wb") as file:
    pickle.dump(clf, file)
    
print("Models saved successfully!")
print("- vectorizer.pkl")
print("- clf.pickle")
```
